```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
from collections import OrderedDict

# CSV 파일 경로
csv_path = '../data/train_updated.csv'

# 이미지 폴더 경로
image_folder_path = '../data/train'

# CSV 파일을 읽기
df = pd.read_csv(csv_path)

# 각 클래스별 이미지를 저장할 딕셔너리
class_images = {}

# 각 클래스별로 하나의 이미지를 선택
for class_label in sorted(df['target'].unique()):
    # 해당 클래스의 첫 번째 이미지 선택
    image_id = df[df['target'] == class_label]['ID'].values[0]
    image_path = os.path.join(image_folder_path, image_id)
    class_images[class_label] = image_path

# 시각화
plt.figure(figsize=(15, 15))
for idx, (class_label, image_path) in enumerate(class_images.items()):
    plt.subplot(4, 5, idx + 1)
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(f'Class {class_label}')
    plt.axis('off')

plt.tight_layout()
plt.show()

#1,3,4,6,7,10,11,12,13,14,15
#0,5,8,9,
#2,16
```

```python
# 클래스별 이미지 개수 계산
class_counts = df['target'].value_counts().sort_index()

# 막대 그래프 시각화
plt.figure(figsize=(8, 3))
class_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.title('Number of Images per Class')
plt.xticks(ticks=class_counts.index, labels=[f'Class {i}' for i in class_counts.index], rotation=45)
plt.grid(axis='y')

plt.show()
```
