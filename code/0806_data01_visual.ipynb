```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
from collections import OrderedDict

# CSV  
csv_path = '../data/train_updated.csv'

#   
image_folder_path = '../data/train'

# CSV  
df = pd.read_csv(csv_path)

#     
class_images = {}

#     
for class_label in sorted(df['target'].unique()):
    #      
    image_id = df[df['target'] == class_label]['ID'].values[0]
    image_path = os.path.join(image_folder_path, image_id)
    class_images[class_label] = image_path

# 
plt.figure(figsize=(15, 15))
for idx, (class_label, image_path) in enumerate(class_images.items()):
    plt.subplot(4, 5, idx + 1)
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(f'Class {class_label}')
    plt.axis('off')

plt.tight_layout()
plt.show()

#1,3,4,6,7,10,11,12,13,14,15
#0,5,8,9,
#2,16
```
